```python
df1 = pd.read_csv("Bengaluru_House_Data.csv")
df1.head()
```

```python
df1["area_type"].value_counts()
```

```python
df2= df1.drop(['area_type','availability','society'],axis=1)
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.drop(["balcony"],axis=1,inplace=True)
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3=df2.dropna()
df3.isnull().sum()
```

```python
df3["size"].unique()
```

```python
def getval(x):
    return int(x.split(" ")[0])

df3["Bhk"]=df3["size"].apply(getval)
df3.head()
```

```python
df3["Bhk"].unique()
```

```python
df3[df3.Bhk>20]
```

```python
df3["total_sqft"].unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
    
```

```python
df3[~df3["total_sqft"].apply(is_float)].head(10)
```

```python
def convert(x):
    token = x.split("-")
    if len(token) == 2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
 convert("2000 - 3000")
```

```python
df4=df3.copy()
df4["total_sqft"]=df4["total_sqft"].apply(convert)
```

```python
df4.loc[30]
```

```python
df5=df4.copy()
df5["cost_per_sqft"]=df5["price"]*100000/df5["total_sqft"]
df5.head()
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: x.strip())
address = df5.groupby("location")["location"].count().sort_values(ascending=False)
address
```

```python
len(address[address<=10])
```

```python
less_locations= address[address<=10]
less_locations
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: "other" if x in less_locations else x )
len(df5.location.unique())
```

```python
df5[df5.total_sqft/df5.Bhk<300].head()
```

```python
df6= df5[~(df5.total_sqft/df5.Bhk<300)]
```

```python
df6.cost_per_sqft.describe()
```

```python
def remove(df):
    df_out =pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.cost_per_sqft)
        s=np.std(subdf.cost_per_sqft)
        reduced_df=subdf[(subdf.cost_per_sqft>m-s)& (subdf.cost_per_sqft<=(m+s))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out

df7=remove(df6)
df7.shape
    
```

```python
def remove_bhk(df):
    exclude_indices = np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk,bhk_df in df.groupby('Bhk'):
            bhk_stats[bhk]={
              'm':np.mean(bhk_df.cost_per_sqft),
              's':np.std(bhk_df.cost_per_sqft),
              'c':bhk_df.shape[0]
            }
        for bhk,bhk_df in df.groupby('Bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['c']>5:
                exclude_indices= np.append(exclude_indices, bhk_df[bhk_df.cost_per_sqft<(stats['m'])].index.values)
    return df.drop(exclude_indices,axis='index')

df8 = remove_bhk(df7)
df8.shape           
```

```python
df8.bath.unique()
```

```python
df8[df8.bath>df8.Bhk+2]
```

```python
df9=df8[~(df8.bath>df8.Bhk+2)]
df9.shape
```

```python
df10= df9.drop(["size","cost_per_sqft"],axis='columns')
df10.head()
```

```python
dummies=pd.get_dummies(df10.location)
dummies.head(5)
```

```python
df11= pd.concat([df10,dummies.drop("other",axis='columns')],axis='columns')
df11.head()
```

```python
df12=df11.drop(["location"],axis=1)
df12.head()
```

```python
df12.shape
```

```python
x=df12.drop(["price"],axis=1)
x
```

```python
y=df12["price"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x.values,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
```

```python
shuffled = ShuffleSplit(n_splits=5,test_size = 0.2,random_state=0)
cross_val_score(LinearRegression(),x,y,cv=shuffled)
```

```python
def predict_cost(location,sqft,bath,bhk):
    loc_index= np.where(x.columns==location)[0][0]
    
    a=np.zeros(len(x.columns))
    a[0]=sqft
    a[1]=bath
    a[2]=bhk
    if loc_index >=0:
        a[loc_index]=1

    return lr.predict([a])[0]


    
```

```python
predict_cost("1st Phase JP Nagar",1000,2,2)
```

```python
predict_cost("1st Phase JP Nagar",1000,3,3)
```

```python
predict_cost("Indira Nagar",1000,3,3)
```
